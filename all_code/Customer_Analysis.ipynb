```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
customer = pd.read_csv("/content/gdrive/MyDrive/Capstone/cleaned_data/customers_clean.csv")
```

```python
customer.head()
```

```python
customer.FN.value_counts().plot(kind='pie', title='FN Distribution', legend=True, autopct='%1.1f%%',)
```

```python
customer.Active.value_counts().plot(kind='pie',labels=['active', 'inactive'], colors = ['blue', 'red'], title='Active Distribution', legend=True, autopct='%1.1f%%')
```

```python
customer.club_member_status.unique() # Categorical Nominal
 
customer.club_member_status.fillna('UNKNOWN').value_counts().plot(kind='pie', colors=['#66b3ff','#ff9999','#99ff99','#ffcc99'], title='Member Status Distribution', legend=True, autopct='%1.1f%%')
```

```python
customer.fashion_news_frequency.unique() # Categorical Ordinal
customer.fashion_news_frequency.fillna('UNKNOWN').value_counts().plot(kind='pie', colors=['#66b3ff','#ff9999','#99ff99','#ffcc99'], title='Member Status Distribution', legend=True, autopct='%1.1f%%')
```

```python
customer.age.value_counts().sort_index().plot(kind='bar', x='age', y = 'count') # Categorical Ordinal
```

```python
len(customer.postal_code.unique()) # Categorical Nominal
customer.postal_code.value_counts().sort_index().plot(kind='bar', x='postal', y = 'count') # Categorical Ordinal
```
